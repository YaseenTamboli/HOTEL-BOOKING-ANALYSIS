```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/my_datasets/hotel_booking.csv")
df.head(10)
```

```python
df.shape
```

```python
df1=df.drop(['name','email','phone-number','credit_card'],axis=1)
```

```python
df1.shape
```

```python
df1.head(10)
```

```python
df1.columns
```

```python
df1.info()
```

```python
df1['reservation_status_date']=pd.to_datetime(df1['reservation_status_date'])
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1.drop(['company','agent'],axis = 1,inplace=True)
```

```python
df1.dropna(inplace=True)
df1.isnull().sum()
```

```python
df1.describe()
```

**Data Analyze**

```python
cancelled_perc=df1['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
plt.figure(figsize=(5,5))
plt.title('Reservation Status Count')
plt.bar(['not_canceled','cancelled'],df1['is_canceled'].value_counts(),color='red',width=0.5)
plt.show()
```

```python

hotels=df1['hotel'].value_counts(normalize=True)
print(hotels)
```

```python
resort_hotel=df1[df1['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df1[df1['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='hotel',hue = 'is_canceled',data=df1,palette = 'Blues')
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
df1['month']=df1['reservation_status_date'].dt.month
plt.figure(figsize=(16,9))
sns.countplot(x='month',hue='is_canceled',data=df1,palette='bright')
plt.title('Reservation Status per month',size=20)
plt.xlabel('reservation status month')
plt.ylabel('number of reservation')
plt.legend(['not_canceled','canceled'])
plt.show()
```

** cancelation rate differnt countries**

```python
canceled_data=df1[df1['is_canceled']==1]
top_10_countries=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_countries,autopct='%0.2f%%',labels=top_10_countries.index)
plt.title('top 10 contries reservation canceled percentage')
plt.show()
```

```python
df1['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df1[df1['is_canceled']==0]
not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend()

```

```python
canceled_df_adr=canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016') & (canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df_adr=not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016') & (not_canceled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=20)
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend(fontsize=15)
plt.show()
```
